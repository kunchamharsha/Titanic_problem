```python
import pandas as pd
```

```python
titanicdata=pd.read_csv('train.csv')
titanicdata.head()
```

```python
Y=titanicdata['Survived']
```

```python
Y
```

```python
fullseries=['Pclass','Sex','Age','Fare']
X=titanicdata.loc[:,fullseries]
```

```python
print(X['Age'])
```

```python
X['Age']=X['Age'].fillna(X.Age.mean())
```

```python
def changesex(row):
    if row.Sex=='male':
        return 1
    elif row.Sex=='female':
        return 0
print(X)
newgender=X.apply(changesex,axis='columns')
print(newgender)
```

```python
X['Sex']=newgender
```

```python
print(X)
```

```python
Y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(multi_class='multinomial', solver='lbfgs').fit(X,Y)
```

```python
clf.predict(X[:5])
```

```python
clf.predict_proba(X[:5]) 
```

```python
Y[:5]
```

```python
clf.score(X,Y)
```
